## 1.加载环境变量

```python
import os

from dotenv import load_dotenv

from LangGraph.envInit import API_KEY

load_dotenv()
```

## 2.定义工具

```python
# 2.定义工具
from langchain.tools import tool


@tool
def getWeather(location: str) -> str:
    """
    Get the weather in a given location.
    Args:
        location: city name or coordinates
    """
    return f"Current weather in {location} is sunny"
```

## 3.创建Agent

```python
from langchain.agents import create_agent
import os

os.environ['API_KEY'] = os.getenv("SILICON_API_KEY")
os.environ['BASE_URL'] = os.getenv("SILICON_BASE_URL_OPEN_AI")

agent = create_agent(
    model="SILICON_BASE_MODEL",
    tools=[getWeather]
)
```

## 4.发起调用



```python
agent.invoke({"input": "今天天气怎么样"})
```
